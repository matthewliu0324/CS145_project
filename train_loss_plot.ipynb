```python
import ast
import matplotlib.pyplot as plt

# Step 1: Read the file and parse the data
file_path = 'training_log.txt'
losses = []
epochs = []
grads = []
lrs = []

with open(file_path, 'r') as file:
    for line in file:
        try:
            log_entry = ast.literal_eval(line.strip())  # Convert string to dictionary
            # Check if the required keys are in the dictionary
            if 'loss' in log_entry and 'epoch' in log_entry:
                losses.append(log_entry['loss'])
                epochs.append(log_entry['epoch'])
                grads.append(log_entry['grad_norm'])
                lrs.append(log_entry['learning_rate'])
        except (ValueError, SyntaxError):
            # Skip the line if it cannot be parsed correctly
            continue
```

```python
# Plot Training loss over epochs
plt.figure(figsize=(10, 6))
plt.plot(epochs, losses, label='Training Loss', c='firebrick')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot training loss for first 200 iterations
plt.figure(figsize=(10, 6))
plt.plot(losses[0:200], label='Training Loss', c='firebrick')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Training Loss for first 200 iterations')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot learning rate over all epochs
plt.figure(figsize=(10, 6))
plt.plot(epochs, lrs, label='Learning Rate', c='firebrick')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot gradient norm over epochs
plt.figure(figsize=(10, 6))
plt.plot(epochs, grads, label='Gradient Norm', c='firebrick')
plt.xlabel('Epoch')
plt.ylabel('Gradient Norm')
plt.title('Gradient Norm over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```
